```python
matrix = np.random.randint(1, 11, size=(2, 3))
print("Matrix:")
print(matrix)
```

```python
data_type = matrix.dtype
print('\nData Type of Matrix:' , data_type)
```

```python
mean_value = np.mean(matrix)
median_value = np.median(matrix)
print(mean_value)
print(median_value)
```

```python
flat_matrix = matrix.flatten()
print(flat_matrix)
```

```python
unique, counts = np.unique(flat_matrix, return_counts = True)
print(unique, counts)
```

```python
mode_index = np.argmax(counts)
print(mode_index)
```

```python
mode_value = unique[mode_index]
print(mode_value)
```

```python
matrix_addition = matrix + matrix
print("\nmatrix addition:")
print(matrix_addition)
```

```python
matrix_subtraction = matrix - matrix
print("\nmatrix subtraction:")
print(matrix_subtraction)
```

```python
matrix_transpose = matrix.T
print("\nmatrix_transpose:")
print(matrix_transpose)
```

```python
matrix_scaler_multiplication = matrix*3
print("\nmatrix scaler multiplication:(matrix*3)")
print(matrix_scaler_multiplication)
```
